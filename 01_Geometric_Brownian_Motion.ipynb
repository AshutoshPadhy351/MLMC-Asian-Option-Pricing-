```python
import numpy as np
import matplotlib.pyplot as plt
```

### Stochastic Differential Equation (SDE):
$$dS_t = \mu S_t dt + \sigma S_t dW_t$$

### Explicit Expression:
$$S_t = S_0 e^{ (\mu - \frac{\sigma^2}{2})t + \sigma W_t }$$

### Other Forms:
For the log-normal distribution, you would write:
$$\log S_t \sim N(\log S_0 + (\mu - \frac{\sigma^2}{2})t, \sigma^2t)$$

```python
mu = 0.2
n=100
T=1
M=100
S0=100
sigma = 0.2
```

```python
dt=T/n
St = np.exp((mu-0.5*sigma**2)*dt + sigma*np.random.normal(0,np.sqrt(dt),size=(M,n)).T)
St = np.vstack((np.ones(M),St))
St = S0*St.cumprod(axis=0)

time = np.linspace(0,T,n+1)
tt= np.full(shape=(M,n+1),fill_value=time).T
```

```python
plt.plot(tt, St)
plt.xlabel("Years $(t)$")
plt.ylabel("Stock Price $(S_t)$")
plt.title(
    "Realizations of Geometric Brownian Motion\n $dS_t = \mu S_t dt + \sigma S_t dW_t$\n $S_0 = {0}, \mu = {1}, \sigma = {2}$".format(S0, mu, sigma)
)
plt.show()
```
